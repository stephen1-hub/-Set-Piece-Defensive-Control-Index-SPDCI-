```python
key_events = [
    'Clearance',
    'Duel',
    'Shot',
    'Block',
    'Interception',
    'Ball Recovery'
]

def get_first_contact(df, idx):
    for i in range(idx+1, idx+15):
        event = df.iloc[i]

        if event['type.name'] in key_events:
            return event['type.name'], event['team.name']

    return None, None
```

# Load Multiple Match Event Files

```python
import pandas as pd
import requests
```

```python
def load_events(match_id):

    url = f"https://raw.githubusercontent.com/statsbomb/open-data/master/data/events/{match_id}.json"

    data = requests.get(url).json()

    return pd.json_normalize(data)
```

```python
priority = {
    "Shot": 3,
    "Clearance": 2,
    "Block": 2,
    "Interception": 2,
    "Duel": 1,
    "Ball Receipt*": 1,
    "Miscontrol": 1
}

def get_first_contact(df, idx):

    best_event = None
    best_score = 0

    max_index = min(idx + 10, len(df))

    for i in range(idx + 1, max_index):

        event = df.iloc[i]
        etype = event.get("type.name", None)
        team = event.get("team.name", None)

        if etype in priority and team is not None:

            if priority[etype] > best_score:
                best_event = (etype, team)
                best_score = priority[etype]

    return best_event if best_event else (None, None)
```

```python
def classify(event):

    if event in ["Clearance", "Block", "Interception"]:
        return "Win"

    elif event == "Shot":
        return "Danger"

    elif event in ["Duel", "Ball Receipt*", "Miscontrol"]:
        return "Contested"

    return "Other"
```

```python
match_ids = [3764440, 3764661, 3773369, 3773372, 3773377]

results = []

for match_id in match_ids:

    df = load_events(match_id)

    corners = df[
        (df["type.name"] == "Pass") &
        (df["pass.type.name"] == "Corner")
    ]

    for idx in corners.index:

        event, team = get_first_contact(df, idx)

        if event is None or team is None:
            continue

        results.append({
            "match_id": match_id,
            "team": team,
            "event": event,
            "outcome": classify(event)
        })

fc_df = pd.DataFrame(results)
```

```python
all_corners = []

for match_id in match_ids:

    df = load_events(match_id)

    corners = df[
        (df["type.name"] == "Pass") &
        (df["pass.type.name"] == "Corner")
    ].copy()

    all_corners.append(corners)

corners_df = pd.concat(all_corners, ignore_index=True)
```

# Define danger zones

```python
def classify_zone(x, y):

    # 6-yard box
    if x >= 114 and 18 <= y <= 62:
        return "6Y_Box"

    # Penalty spot corridor
    elif 108 <= x < 114 and 18 <= y <= 62:
        return "Penalty_Spot"

    # Wide deliveries
    elif x >= 108 and (y < 18 or y > 62):
        return "Wide_Delivery"

    # Short corners / edge-of-box deliveries
    else:
        return "Edge_Zone"
```

```python
corners["zone"] = corners.apply(
    lambda row: classify_zone(row["end_x"], row["end_y"]),
    axis=1
)
```

```python
def process_match(match_id):

    df = load_events(match_id)

    corners = df[
        (df["type.name"] == "Pass") &
        (df["pass.type.name"] == "Corner")
    ].copy()

    if corners.empty:
        return None

    # coordinates
    corners["end_x"] = corners["pass.end_location"].apply(lambda v: v[0])
    corners["end_y"] = corners["pass.end_location"].apply(lambda v: v[1])

    # zone classification
    corners["zone"] = corners.apply(
        lambda row: classify_zone(row["end_x"], row["end_y"]),
        axis=1
    )

    # get team names present in match
    teams = df["team.name"].dropna().unique()

    if len(teams) != 2:
        return None

    team_a, team_b = teams[0], teams[1]

    # assign defending team
    corners["defending_team"] = corners["team.name"].apply(
        lambda x: team_b if x == team_a else team_a
    )

    weights = {
        "6Y_Box": 4,
        "Penalty_Spot": 3,
        "Wide_Delivery": 2,
        "Edge_Zone": 1
    }

    corners["danger_score"] = corners["zone"].map(weights)

    dhss = corners.groupby("defending_team", as_index=False).agg(
        Corners_Faced=("danger_score", "count"),
        Total_Danger=("danger_score", "sum"),
        Avg_Danger=("danger_score", "mean")
    )

    dhss["DHSS"] = (4 - dhss["Avg_Danger"]) / 4 * 100
    dhss["match_id"] = match_id

    return dhss
```

```python
all_results = []

for m in match_ids:
    result = process_match(m)

    if result is not None:
        all_results.append(result)

league_df = pd.concat(all_results, ignore_index=True)

league_df.head(20)
```

```python
league_summary = (
    league_df.groupby("defending_team", as_index=False)
    .agg(
        Matches=("match_id", "nunique"),
        Corners_Faced=("Corners_Faced", "sum"),
        Total_Danger=("Total_Danger", "sum")
    )
)

league_summary["Avg_Danger"] = (
    league_summary["Total_Danger"] /
    league_summary["Corners_Faced"]
)

league_summary["DHSS"] = (
    (4 - league_summary["Avg_Danger"]) / 4
) * 100

league_summary = league_summary.sort_values(
    "DHSS",
    ascending=False
)

league_summary
```

```python
def process_match_sbcm(match_id):

    df = load_events(match_id)

    # get corners
    corners = df[
        (df["type.name"] == "Pass") &
        (df["pass.type.name"] == "Corner")
    ].copy()


    if corners.empty:
        return None


    teams = df["team.name"].dropna().unique()

    if len(teams) != 2:
        return None


    team_a, team_b = teams[0], teams[1]


    # TRUE first contact events
    first_contact_events = [
        "Clearance",
        "Duel",
        "Block",
        "Interception",
        "Miscontrol"
    ]


    control_events = [
        "Pass",
        "Carry",
        "Ball Recovery",
        "Interception"
    ]


    records = []


    for _, corner in corners.iterrows():

        corner_idx = corner["index"]

        attacking_team = corner["team.name"]

        defending_team = (
            team_b if attacking_team == team_a else team_a
        )


        # find first contest
        first_contact = df[

            (df["index"] > corner_idx) &

            (df["index"] <= corner_idx + 10) &

            (df["type.name"].isin(first_contact_events))

        ].head(1)


        if first_contact.empty:
            continue


        first_idx = first_contact.iloc[0]["index"]


        # find second phase control
        second_phase = df[

            (df["index"] > first_idx) &

            (df["index"] <= first_idx + 20) &

            (df["type.name"].isin(control_events))

        ].head(1)


        if second_phase.empty:
            continue


        records.append({

            "attacking_team": attacking_team,

            "defending_team": defending_team,

            "first_action": first_contact.iloc[0]["type.name"],

            "first_contact_team": first_contact.iloc[0]["team.name"],

            "second_action": second_phase.iloc[0]["type.name"],

            "second_ball_winner": second_phase.iloc[0]["team.name"]

        })


    if len(records) == 0:
        return None


    second_ball_df = pd.DataFrame(records)


    second_ball_df["defense_controlled"] = (

        second_ball_df["defending_team"]

        ==

        second_ball_df["second_ball_winner"]

    )


    sbcm = second_ball_df.groupby(
        "defending_team",
        as_index=False
    ).agg(

        Second_Ball_Situations=(
            "defense_controlled",
            "count"
        ),

        Second_Balls_Controlled=(
            "defense_controlled",
            "sum"
        )

    )


    sbcm["SBCM"] = (

        sbcm["Second_Balls_Controlled"]

        /

        sbcm["Second_Ball_Situations"]

    ) * 100


    sbcm["match_id"] = match_id


    return sbcm
```

# Create FCCI match function

```python
def process_match_fcci(match_id):

    df = load_events(match_id)

    corners = df[
        (df["type.name"] == "Pass") &
        (df["pass.type.name"] == "Corner")
    ].copy()


    if corners.empty:
        return None


    teams = df["team.name"].dropna().unique()

    if len(teams) != 2:
        return None


    team_a, team_b = teams[0], teams[1]


    first_contact_events = [
        "Clearance",
        "Duel",
        "Block",
        "Interception",
        "Miscontrol"
    ]


    records = []


    for _, corner in corners.iterrows():

        corner_idx = corner["index"]

        attacking_team = corner["team.name"]

        defending_team = (
            team_b if attacking_team == team_a else team_a
        )


        first_contact = df[
            (df["index"] > corner_idx)
            &
            (df["index"] <= corner_idx + 10)
            &
            (df["type.name"].isin(first_contact_events))
        ].head(1)


        if first_contact.empty:
            continue


        records.append({

            "attacking_team": attacking_team,

            "defending_team": defending_team,

            "first_action": first_contact.iloc[0]["type.name"],

            "first_contact_team": first_contact.iloc[0]["team.name"]

        })


    if len(records) == 0:
        return None


    fc = pd.DataFrame(records)


    fc["defense_won_contact"] = (

        fc["defending_team"]

        ==

        fc["first_contact_team"]

    )


    result = fc.groupby(
        "defending_team",
        as_index=False
    ).agg(

        First_Contact_Situations=(
            "defense_won_contact",
            "count"
        ),

        First_Contacts_Won=(
            "defense_won_contact",
            "sum"
        )

    )


    result["FCCI"] = (

        result["First_Contacts_Won"]

        /

        result["First_Contact_Situations"]

    ) * 100


    result["match_id"] = match_id


    return result
```
